```python
from oceanicospy.models.swanpy import Initializer
from oceanicospy.models.swanpy.preprocess import GridMaker, BathyMaker
from oceanicospy.gis import XYZFormatSpec
```

```python
path_case = '../../data/model_runs/temp_swan_case/'

dict_ini_data = dict(
    name            = 'Stationary case',
    case_number     = 1,
    case_description= 'Example',
    stat_id         = 1,
    number_domains  = 1
)
```

```python
case = Initializer(
    root_path    = path_case,
    dict_ini_data= dict_ini_data,
)

case.create_folders()    # creates all project folders (data/, output/, etc.)
case.replace_ini_data()     # writes run.swn from the STAT template
```

```python
DOMAIN_ID = 1   # domain identifier (1-indexed)

case_grid = GridMaker(
    init         = case,
    domain_number= DOMAIN_ID
)
```

```python
case_bathy = BathyMaker(
    init         = case,
    domain_number= DOMAIN_ID,
)
```

```python
case_bathy.convert_xyz2asc(xyz_filepath='bat_sai_10m_9377.xyz', dx=100, dy=100, xyz_format=XYZFormatSpec(delimiter=' ',has_header=False))
print(case_bathy.bathy_info)
```
